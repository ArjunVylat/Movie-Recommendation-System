Data Loading

```python
import pandas as pd

columns = ['user_id', 'item_id', 'rating', 'timestamp']
ratings_df = pd.read_csv(r'ml-100k\u.data', sep='\t', names=columns)

ratings_df
```

```python
movie_info_columns = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movie_df = pd.read_csv(r'ml-100k\u.item', sep = '|', names = movie_info_columns, encoding= 'latin-1')

movie_df
```

```python
df = pd.merge(ratings_df, movie_df, left_on= 'item_id', right_on= 'movie id')

df
```

Data Preprocessing

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['movie title'] = le.fit_transform(df['movie title'])
df['release date'] = le.fit_transform(df['release date'])
df['IMDb URL'] = le.fit_transform(df['IMDb URL'])

df
```

```python
df = df.drop('video release date', axis = 1)

df
```
