Feature Engineering

```python
from sklearn.feature_extraction.text import TfidfVectorizer


genre_cols = ['unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
genre_df = movie_df[genre_cols]
genre_df['genre_text'] = genre_df.apply(lambda x : ' '.join(genre for genre in genre_cols if x[genre] == 1), axis = 1)

print(genre_df['genre_text'])

tfidf = TfidfVectorizer(stop_words= 'english', max_features=100)
tfidf_matrix = tfidf.fit_transform(genre_df['genre_text'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns= tfidf.get_feature_names_out())

print(tfidf_matrix)
```

Trainig the Model using Collaborative Filtering and Content Based Filtering

```python
user_item_matrix = ratings_df.pivot_table(index= 'user_id', columns= 'item_id', values= 'rating')

user_item_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

user_similarity = cosine_similarity(user_item_matrix.fillna(0))

user_similarity
```

```python
item_similarity = cosine_similarity(tfidf_matrix)

item_similarity
```

```python
import numpy as np

def hybrid_recommendations(user_id, num_recommendations = 5):
    user_similarities = user_similarity[user_id]
    cf_recommendations = np.argsort(user_similarities)[: : -1][1 : ]

    user_rated_items = user_item_matrix.loc[user_id].notna()
    item_similarities = item_similarity[user_rated_items]
    cbf_recommendations = np.argsort(item_similarities.sum(axis = 0))[: : -1]

    hybrid_recommendations = set(cf_recommendations).union(set(cbf_recommendations))

    return list(hybrid_recommendations)[: num_recommendations]
```

```python
user_id_to_test = 5
recommendations = hybrid_recommendations(user_id_to_test)
recommended_movies = [item_id for item_id in recommendations]

print(recommended_movies)

print(f'Hybrid Recommendations for User {user_id_to_test} : ')

for i, movie_id in enumerate(recommendations):
    movie_title = movie_df['movie_title'].iloc[movie_id]
    print(f'{i + 1}. {movie_title}')
```

Evalauting Model Using Precision and Recall Metrics

```python
def precision_recall_at_k(predictions, k = 5, threshold = 3.5):
    predictions = np.array(predictions)
    top_k = (-predictions).argsort()
    precision = []
    recall = []

    for i in range(len(top_k)):
                    relavent_items = np.sum(predictions[top_k[i]] >= threshold)

                    if np.isscalar(predictions[top_k[i]]):
                            recommended_items = 1
                            predicted_items = [predictions[top_k[i]]]
                    else:
                            recommended_items = min(k, len(predictions[top_k[i]]))
                            predicted_items = predictions[top_k[i]][ : recommended_items].tolist()
                    
                    true_positives = np.sum(np.array(predicted_items) >= threshold)

                    precision_i = true_positives / recommended_items if recommended_items > 0 else 0
                    recall_i = true_positives / relavent_items if relavent_items > 0 else 0

                    precision.append(precision_i)
                    recall.append(recall_i)
    
    avg_precision = np.mean(precision)
    avg_recall = np.mean(recall)

    return avg_precision, avg_recall
```

```python
avg_precision, avg_recall = precision_recall_at_k(recommendations, k = 5, threshold=3.5)

print(f'Average Precision : {avg_precision : .2f}')
print(f'Average Recall : {avg_recall : .2f}')
```
