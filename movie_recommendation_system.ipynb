Data Loading

```python
import pandas as pd

columns = ['user id', 'item id', 'rating', 'timestamp']
ratings_df = pd.read_csv(r'ml-100k\u.data', sep='\t', names=columns)

ratings_df
```

```python
movie_info_columns = ['movie id', 'movie title', 'release date', 'video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
movie_df = pd.read_csv(r'ml-100k\u.item', sep = '|', names = movie_info_columns, encoding= 'latin-1')

movie_df
```

```python
movie_df['num_genres'] = movie_df.iloc[:, 6 :].sum(axis= 1)

movie_df.head()
```

```python
df = pd.merge(ratings_df, movie_df, left_on= 'item id', right_on= 'movie id')

df.head()
```

Data Preprocessing

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['movie title_encoded'] = le.fit_transform(df['movie title'])
df['release date'] = le.fit_transform(df['release date'])
df['IMDb URL'] = le.fit_transform(df['IMDb URL'])

df
```

```python
df = df.drop('video release date', axis = 1)

df
```

```python
df.isnull().sum()
```

Feature Engineering

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words= 'english', max_features= 100)
tfidf_matrix = tfidf.fit_transform(movie_df['movie title'])
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns= tfidf.get_feature_names_out())

print(tfidf_df)
```
